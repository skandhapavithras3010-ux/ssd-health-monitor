```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/ssd_data.csv")
df.head()
```

```python
FEATURES = [
    "temperature",
    "wear_level",
    "power_on_hours",
    "write_cycles",
    "ECC_error_rate",
    "bad_blocks"
]

X = df[FEATURES]
```

```python
y_class = df["health_status"]
y_rul = df["RUL"]
```

```python
# Manual human-friendly encoding
mapping = {
    "Healthy": 0,
    "Warning": 1,
    "Critical": 2
}

y_class_enc = y_class.map(mapping)

print(mapping)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, yc_train, yc_test, yr_train, yr_test = train_test_split(
    X,
    y_class_enc,
    y_rul,
    test_size=0.2,
    random_state=42,
    stratify=y_class_enc
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_s = scaler.fit_transform(X_train)

X_test_s = scaler.transform(X_test)
```
